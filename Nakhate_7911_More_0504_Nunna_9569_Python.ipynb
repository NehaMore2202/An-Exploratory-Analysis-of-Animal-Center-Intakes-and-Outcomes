### Task 1C: 4 points
1. Plot the Age Intake column and observe the distribution.
2. Group certain values into categories,
- Less than 1 week, Less than 1 month, Less than 1 year
- Drop inappropriate negative valued years rows.
3. Plot the Column to again see the new distributuions.

Tip: consider 0 years as less than 1 year.

```python
plt.figure(figsize=(50, 6))
sns.histplot(animal_data['Age upon Intake'], bins=20, kde=True, color='blue')
plt.title('Distribution of Age Intake')
plt.xlabel('Age Intake')
plt.ylabel('Frequency')
plt.show()
```

```python
animal_data['Age upon Intake']
```

```python
def parse_to_timedelta(date_str):
    quantity, unit = date_str.split()
    quantity = int(quantity)
    if unit=='days' or unit=='day':
        return timedelta(days=quantity)
    elif unit == 'week' or unit == 'weeks':
        return timedelta(days=quantity*7)
    elif unit == 'month' or unit == 'months':
        return timedelta(days=quantity * 30)
    elif unit == 'year' or unit == 'years':
        return timedelta(days=quantity * 365)
    else:
        raise ValueError(f'Unknown unit: {unit}')

animal_data['Age upon Intake'] = animal_data['Age upon Intake'].apply(parse_to_timedelta)

print(animal_data)
```

```python
Age=animal_data['Age upon Intake']
```

```python
def categorize_age(age):
        td = age.days
        if int(td) < 7:
            return 'Less than a week'
        elif int(td) < 31:
            return 'Less than a month'
        elif int(td) < 365:
            return 'Less than a year'
        else:
          return 'Other'
age_categories = [categorize_age(age) for age in Age]
df1 = pd.DataFrame({'Category': age_categories})
```

```python
animal_data['Age Category']=df1
```

```python
AgeCat = animal_data['Age Category'].value_counts()
```

```python
plt.figure(figsize=(20, 6))
plt.bar(AgeCat.index,AgeCat.values)
plt.title('Distribution of Age Intake')
plt.xlabel('Age Intake')
plt.ylabel('Frequency')
plt.show()
```

### Task 1E: 2 points

### Task 1F: 4 points
Display horizontal bar graph for Top 10 Locations where animals were found at night time that were sick.

Note: night time = 7pm to 12am

```python
sick_animals = animal_data[(animal_data['DateTime'].dt.hour >= 19) & (animal_data['DateTime'].dt.hour <= 23) & (animal_data['Intake Condition'] == 'Sick')]
location_counts = sick_animals['Found Location'].value_counts()
top_10_locations_Sick = location_counts.head(10)
```
